Top 10 countries with highest Malaria Cases 2017, 2016 and 2015 

```python
#2017
# Filter the data for the year 2017
data_2017 = malaria_df[malaria_df['yr'] == 2017]

# Sort the data by the number of malaria cases reported in descending order
sorted_data_2017 = data_2017.sort_values(by='malaria_cases', ascending=False)

# Select the top 10 countries with the highest number of malaria cases reported in 2017
top_10_countries_2017 = sorted_data_2017.head(10)

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries_2017['cntry_name'], top_10_countries_2017['malaria_cases'], color='blue')
plt.xlabel('Country')
plt.ylabel('Malaria Cases Reported (2017)')
plt.title('Top 10 Countries with the Highest Malaria Cases Reported in 2017')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#2016
# Filter the data for the year 2017
data_2016 = malaria_df[malaria_df['yr'] == 2016]

# Sort the data by the number of malaria cases reported in descending order
sorted_data_2016 = data_2016.sort_values(by= 'malaria_cases', ascending=False)

# Select the top 10 countries with the highest number of malaria cases reported in 2016
top_10_countries_2016 = sorted_data_2016.head(10)

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries_2016['cntry_name'], top_10_countries_2016['malaria_cases'], color='blue')
plt.xlabel('Country')
plt.ylabel('Malaria Cases Reported (2017)')
plt.title('Top 10 Countries with the Highest Malaria Cases Reported in 2016')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Assuming you have loaded the DataFrame "malaria_df" and renamed the columns with short forms

# Calculate the average data for each country
average_data = malaria_df.groupby('cntry_name').mean().reset_index()

# Create a pie chart using Plotly Express
fig = px.pie(average_data, values='safe_drinking_water(% Rural)', names='cntry_name', 
             title='Average People using Safely Managed Drinking Water Services (Rural vs. Urban)',
             hover_data=['safe_drinking_water(% Rural)', 'safe_drinking_water(% Urban)'])

# Add the percentage of urban water services to the hover information
fig.update_traces(textinfo='percent+label')

# Show the chart
fig.show()

# Find the country with the best and worst safe drinking water services
best_country = average_data.loc[average_data['safe_drinking_water(% Rural)'].idxmax(), 'cntry_name']
worst_country = average_data.loc[average_data['safe_drinking_water(% Rural)'].idxmin(), 'cntry_name']

print(f"The country with the best safe drinking water services is: {best_country}")
print(f"The country with the worst safe drinking water services is: {worst_country}")

```
