```python
#creating a dictionary with the shorter forms for the columns
short_forms = {
    'Country Name': 'cntry_name',
    'Year': 'yr',
    'Country Code': 'cntry_code',
    'Incidence of malaria (per 1,000 population at risk)': 'malaria_incidence',
    'Malaria cases reported': 'malaria_cases',
    'Use of insecticide-treated bed nets (% of under-5 population)': 'ITNs_usage(% <5)',
    'Children with fever receiving antimalarial drugs (% of children under age 5 with fever)': 'antimalarial_Rx(% <5)',
    'Intermittent preventive treatment (IPT) of malaria in pregnancy (% of pregnant women)': 'IPT_in_pregnancy(% Pregnant)',
    'People using safely managed drinking water services (% of population)': 'safe_drinking_water(% Total)',
    'People using safely managed drinking water services, rural (% of rural population)': 'safe_drinking_water(% Rural)',
    'People using safely managed drinking water services, urban (% of urban population)': 'safe_drinking_water(% Urban)',
    'People using safely managed sanitation services (% of population)': 'safe_sanitation(% Total)',
    'People using safely managed sanitation services, rural (% of rural population)': 'safe_sanitation(% Rural)',
    'People using safely managed sanitation services, urban  (% of urban population)': 'safe_sanitation(% Urban)',
    'Rural population (% of total population)': 'rural_pop(% Total)',
    'Rural population growth (annual %)': 'rural_pop_growth(%)',
    'Urban population (% of total population)': 'urban_pop(% Total)',
    'Urban population growth (annual %)': 'urban_pop_growth(%)',
    'People using at least basic drinking water services (% of population)': 'basic_drinking_water(% Total)',
    'People using at least basic drinking water services, rural (% of rural population)': 'basic_drinking_water(% Rural)',
    'People using at least basic drinking water services, urban (% of urban population)': 'basic_drinking_water(% Urban)',
    'People using at least basic sanitation services (% of population)': 'basic_sanitation(% Total)',
    'People using at least basic sanitation services, rural (% of rural population)': 'basic_sanitaion(% Rural)',
    'People using at least basic sanitation services, urban  (% of urban population)': 'basic_sanitation(% Urban)',
    'latitude': 'lat',
    'longitude': 'long',
    'geometry': 'geo'
}

# Rename the columns using the short forms
malaria_df.rename(columns=short_forms, inplace=True)
```

```python
# Handling the missing values
malaria_df.columns
```

```python
#filling the missing values in the columns with the mean
malaria_df['rural_pop(% Total)'].fillna(malaria_df['rural_pop(% Total)'].mean(), inplace=True)

malaria_df['urban_pop(% Total)'].fillna(malaria_df['urban_pop(% Total)'].mean(), inplace=True)

malaria_df['rural_pop_growth(%)'].fillna(malaria_df['rural_pop_growth(%)'].mean(), inplace=True)

malaria_df['urban_pop_growth(%)'].fillna(malaria_df['urban_pop_growth(%)'].mean(), inplace=True)

malaria_df['basic_drinking_water(% Total)'].fillna(malaria_df['basic_drinking_water(% Total)'].mean(), inplace=True)

malaria_df['basic_drinking_water(% Rural)'].fillna(malaria_df['basic_drinking_water(% Rural)'].mean(), inplace=True)

malaria_df['basic_sanitaion(% Rural)'].fillna(malaria_df['basic_sanitaion(% Rural)'].mean(), inplace=True)

malaria_df['basic_sanitation(% Urban)'].fillna(malaria_df['basic_sanitation(% Urban)'].mean(), inplace=True)

malaria_df['basic_drinking_water(% Urban)'].fillna(malaria_df['basic_drinking_water(% Urban)'].mean(), inplace=True)

malaria_df['basic_sanitation(% Total)'].fillna(malaria_df['basic_sanitation(% Total)'].mean(), inplace=True)

malaria_df['malaria_incidence'].fillna(malaria_df['malaria_incidence'].mean(), inplace=True)

malaria_df['malaria_cases'].fillna(malaria_df['malaria_cases'].mean(), inplace=True)
```

```python
# replace the missing values in the columns with 0
malaria_df['ITNs_usage(% <5)'].fillna(0, inplace=True)

malaria_df['antimalarial_Rx(% <5)'].fillna(0, inplace=True)

malaria_df['IPT_in_pregnancy(% Pregnant)'].fillna(0, inplace=True)

malaria_df['safe_drinking_water(% Total)'].fillna(0, inplace=True)

malaria_df['safe_drinking_water(% Rural)'].fillna(0, inplace=True)

malaria_df['safe_drinking_water(% Urban)'].fillna(0, inplace=True)

malaria_df['safe_sanitation(% Total)'].fillna(0, inplace=True)

malaria_df['safe_sanitation(% Rural)'].fillna(0, inplace=True)

malaria_df['safe_sanitation(% Urban)'].fillna(0, inplace=True)

```

### Top 10 countries with highest Malaria Cases 2017, 2016 and 2015 

```python
#2017
# Filter the data for the year 2017
data_2017 = malaria_df[malaria_df['yr'] == 2017]

# Sort the data by the number of malaria cases reported in descending order
sorted_data_2017 = data_2017.sort_values(by='malaria_cases', ascending=False)

# Select the top 10 countries with the highest number of malaria cases reported in 2017
top_10_countries_2017 = sorted_data_2017.head(10)

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries_2017['cntry_name'], top_10_countries_2017['malaria_cases'], color='blue')
plt.xlabel('Country')
plt.ylabel('Malaria Cases Reported (2017)')
plt.title('Top 10 Countries with the Highest Malaria Cases Reported in 2017')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#2016
# Filter the data for the year 2017
data_2016 = malaria_df[malaria_df['yr'] == 2016]

# Sort the data by the number of malaria cases reported in descending order
sorted_data_2016 = data_2016.sort_values(by= 'malaria_cases', ascending=False)

# Select the top 10 countries with the highest number of malaria cases reported in 2016
top_10_countries_2016 = sorted_data_2016.head(10)

# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries_2016['cntry_name'], top_10_countries_2016['malaria_cases'], color='blue')
plt.xlabel('Country')
plt.ylabel('Malaria Cases Reported (2017)')
plt.title('Top 10 Countries with the Highest Malaria Cases Reported in 2016')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Linear graphs showing reported cases of West African countries over time 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming malaria_data is the DataFrame containing your Malaria dataset

# Step 1: Define the list of countries in Western Africa
western_african_countries = ['Nigeria', 'Ghana', 'Senegal', 'Mali', 'Ivory Coast', 'Burkina Faso']

# Step 2: Filter data for specific countries in Western Africa
data_western_africa = malaria_df[malaria_df['cntry_name'].isin(western_african_countries)]

# Step 3: Create line charts
plt.figure(figsize=(12, 6))

# Line chart for specific countries in Western Africa
for country in western_african_countries:
    data_country = data_western_africa[data_western_africa['cntry_name'] == country]
    plt.plot(data_country['yr'], data_country['malaria_cases'], label=country)

plt.xlabel('Year')
plt.ylabel('Malaria cases reported')
plt.title('Time Series: Reported Cases of Malaria for Specific Countries in Western Africa')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
