```python
# Average incidence of malaria in each region in 2009 and 2010
avg_incidence_western_africa_2009 = data_2009[data_2009['cntry_name'].isin(western_african_countries)]['malaria_incidence'].mean()
avg_incidence_eastern_africa_2009 = data_2009[data_2009['cntry_name'].isin(east_african_countries)]['malaria_incidence'].mean()
avg_incidence_southern_africa_2009 = data_2009[data_2009['cntry_name'].isin(southern_african_countries)]['malaria_incidence'].mean()
avg_incidence_northern_africa_2009 = data_2009[data_2009['cntry_name'].isin(northern_african_countries)]['malaria_incidence'].mean()
avg_incidence_central_africa_2009 = data_2009[data_2009['cntry_name'].isin(central_african_countries)]['malaria_incidence'].mean()

avg_incidence_western_africa_2010 = data_2010[data_2010['cntry_name'].isin(western_african_countries)]['malaria_incidence'].mean()
avg_incidence_eastern_africa_2010 = data_2010[data_2010['cntry_name'].isin(east_african_countries)]['malaria_incidence'].mean()
avg_incidence_southern_africa_2010 = data_2010[data_2010['cntry_name'].isin(southern_african_countries)]['malaria_incidence'].mean()
avg_incidence_northern_africa_2010 = data_2010[data_2010['cntry_name'].isin(northern_african_countries)]['malaria_incidence'].mean()
avg_incidence_central_africa_2010 = data_2010[data_2010['cntry_name'].isin(central_african_countries)]['malaria_incidence'].mean()


regions = ['Western Africa', 'Eastern Africa', 'Southern Africa', 'Northern Africa', 'Central Africa']
years = ['2009', '2010']
avg_incidence = [
    [avg_incidence_western_africa_2009, avg_incidence_western_africa_2010],
    [avg_incidence_eastern_africa_2009, avg_incidence_eastern_africa_2010],
    [avg_incidence_southern_africa_2009, avg_incidence_southern_africa_2010],
    [avg_incidence_northern_africa_2009, avg_incidence_northern_africa_2010],
    [avg_incidence_central_africa_2009, avg_incidence_central_africa_2010]
]

# Plotting the grouped bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.4
index = range(len(regions))
for i, year in enumerate(years):
    plt.bar([pos + i * bar_width for pos in index], [inc[i] for inc in avg_incidence], bar_width, label=year)

plt.xlabel('Region')
plt.ylabel('Average Incidence of Malaria')
plt.title('Average Incidence of Malaria in Each Region (2009-2010)')
plt.xticks([pos + bar_width / 2 for pos in index], regions)
plt.legend(title='Year', loc='upper right')
plt.show()
```
