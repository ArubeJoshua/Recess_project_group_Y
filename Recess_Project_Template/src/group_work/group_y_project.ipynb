DATA PREPARATION AND CLEANING(BY LUBEMBE COLIN WAFULA)

```python
# Handling the missing values
malaria_df.columns
```

```python
#filling the missing values in the columns with the mean
malaria_df['Rural population (% of total population)'].fillna(malaria_df['Rural population (% of total population)'].mean(), inplace=True)

malaria_df['Urban population (% of total population)'].fillna(malaria_df['Urban population (% of total population)'].mean(), inplace=True)

malaria_df['Rural population growth (annual %)'].fillna(malaria_df['Rural population growth (annual %)'].mean(), inplace=True)

malaria_df['Urban population growth (annual %)'].fillna(malaria_df['Urban population growth (annual %)'].mean(), inplace=True)

malaria_df['People using at least basic drinking water services (% of population)'].fillna(malaria_df['People using at least basic drinking water services (% of population)'].mean(), inplace=True)

malaria_df['People using at least basic drinking water services, rural (% of rural population)'].fillna(malaria_df['People using at least basic drinking water services, rural (% of rural population)'].mean(), inplace=True)

malaria_df['People using at least basic sanitation services, rural (% of rural population)'].fillna(malaria_df['People using at least basic sanitation services, rural (% of rural population)'].mean(), inplace=True)

malaria_df['People using at least basic sanitation services, urban  (% of urban population)'].fillna(malaria_df['People using at least basic sanitation services, urban  (% of urban population)'].mean(), inplace=True)

malaria_df['People using at least basic drinking water services, urban (% of urban population)'].fillna(malaria_df['People using at least basic drinking water services, urban (% of urban population)'].mean(), inplace=True)

malaria_df['People using at least basic sanitation services (% of population)'].fillna(malaria_df['People using at least basic sanitation services (% of population)'].mean(), inplace=True)
```

```python
# replace the missing values in the columns with 0
malaria_df['Use of insecticide-treated bed nets (% of under-5 population)'].fillna(0, inplace=True)

malaria_df['Children with fever receiving antimalarial drugs (% of children under age 5 with fever)'].fillna(0, inplace=True)

malaria_df['Intermittent preventive treatment (IPT) of malaria in pregnancy (% of pregnant women)'].fillna(0, inplace=True)

malaria_df['People using safely managed drinking water services (% of population)'].fillna(0, inplace=True)

malaria_df['People using safely managed drinking water services, rural (% of rural population)'].fillna(0, inplace=True)

malaria_df['People using safely managed drinking water services, urban (% of urban population)'].fillna(0, inplace=True)

malaria_df['People using safely managed sanitation services (% of population)'].fillna(0, inplace=True)

malaria_df['People using safely managed sanitation services, rural (% of rural population)'].fillna(0, inplace=True)

malaria_df['People using safely managed sanitation services, urban  (% of urban population)'].fillna(0, inplace=True)

malaria_df['Incidence of malaria (per 1,000 population at risk)'].fillna(0, inplace=True)

malaria_df['Malaria cases reported'].fillna(0, inplace=True)
```
